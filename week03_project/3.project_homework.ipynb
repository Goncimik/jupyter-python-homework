            {"isim": "Veri Bilimi 101", "yazar": "Ali", "tur": "Bilim", "satis": 1200, "yil": 2021},
            {"isim": "Python ile Yapay Zeka", "yazar": "Ayşe", "tur": "Bilim", "satis": 950, "yil":2020},
            {"isim": "İstatistik Temelleri", "yazar": "Ali", "tur": "Akademik", "satis": 700, "yil": 2019},
            {"isim": "Makine Öğrenmesi", "yazar": "Can", "tur": "Bilim", "satis": 1800, "yil": 2022},
            {"isim": "Veri Görselleştirme", "yazar": "Deniz", "tur": "Sanat", "satis": 400, "yil": 2018},
            {"isim": "Matematiksel Modelleme", "yazar": "Ali", "tur": "Akademik", "satis": 1500,"yil": 2021},
            {"isim": "Bilgi Toplumu", "yazar": "Ayşe", "tur": "Sosyal", "satis": 600, "yil": 2022}
            ]

""""
# 📌Yapılması Gerekenler:
"""
📝 1. Fonksiyon Yazma:
        o en_cok_satan(kitaplar) → En çok satan kitabın bilgilerini döndürsün.
        o yazar_satislari(kitaplar) → Her yazarın toplam satışını bir sözlük olarak döndürsün.
📦 2. Liste ve Küme İşlemleri:
        o Tüm kitap türlerini (tur) küme halinde çıkarın (tekrar eden türler olmadan).
        o Satış adedi 1000’den fazla olan kitapların isimlerini bir listede toplayın.
⚡ 3. Lambda / Filter / Map Kullanımı:
        o filter ile 2020’den sonra çıkan kitapları süzün.
        o map ile tüm satış adetlerini %10 artırılmış şekilde yeni bir listeye aktarın.
        o sorted + lambda ile kitapları satış miktarına göre azalan şekilde sıralayın
🗂️ 4. İstatistiksel Analiz:
        o Ortalama satış adedini bulun.
        o En çok satış yapan türü bulun.
        o Satışların standart sapmasını hesaplamak için statistics modülünü kullanın.
📈 5. Ekstra (Zorlayıcı): Train/Test Simülasyonu
        o Kitap listesini rastgele %70 eğitim (train), %30 test verisine ayırın (random.sample).
        o Eğitim verisinden yazarların ortalama satışını hesaplayın.
        o Test verisinde, hangi kitapların satışlarının bu ortalamanın üzerinde olduğunu kontrol edin.
"""

# 📌Beklenen Çıktı Örneği
"""
• En çok satan kitap: "Makine Öğrenmesi"
• Yazar satışları: {"Ali": 3400, "Ayşe": 1550, "Can": 1800, "Deniz": 400}
• Türler: {"Bilim", "Akademik", "Sanat", "Sosyal"}
• 1000’den fazla satan kitaplar: ["Veri Bilimi 101", "Makine Öğrenmesi", "Matematiksel Modelleme"]
• Ortalama satış: 1021.4
• Standart sapma: ~480.2
• Eğitim/ Test ayırımı sonrası analiz:
"""

# ⚡Analizde Sizden Beklenen:
"""
• Veriyi uygun oranlarda ayırın (örneğin %70 train, %30 test).
• Train seti üzerinde basit analizler yapın (ortalama, sıklık, vs.).
• Test seti üzerinde aynı analizleri tekrarlayın.
• Sonuçları karşılaştırın ve kısa yorum ekleyin.
"""

```python
# 📌 Gerekli Kütüphaneler
import random
import statistics
```

```python
# 📌 Veri Seti
kitaplar = [
        {"isim": "Veri Bilimi 101", "yazar": "Ali", "tur": "Bilim", "satis": 1200, "yil": 2021},
        {"isim": "Python ile Yapay Zeka", "yazar": "Ayşe", "tur": "Bilim", "satis": 950, "yil":2020},
        {"isim": "İstatistik Temelleri", "yazar": "Ali", "tur": "Akademik", "satis": 700, "yil": 2019},
        {"isim": "Makine Öğrenmesi", "yazar": "Can", "tur": "Bilim", "satis": 1800, "yil": 2022},
        {"isim": "Veri Görselleştirme", "yazar": "Deniz", "tur": "Sanat", "satis": 400, "yil": 2018},
        {"isim": "Matematiksel Modelleme", "yazar": "Ali", "tur": "Akademik", "satis": 1500,"yil": 2021},
        {"isim": "Bilgi Toplumu", "yazar": "Ayşe", "tur": "Sosyal", "satis": 600, "yil": 2022}
        ]
```

```python
# 📌 1. Fonksiyonlar
def en_cok_satan(kitap_listesi):
    return max(kitap_listesi, key=lambda k: k['satis'])

def yazar_satislari(kitap_listesi):
    satis_dict = {}
    for k in kitap_listesi:
        satis_dict[k['yazar']] = satis_dict.get(k['yazar'], 0) + k['satis']
    return satis_dict

def tur_satislari(kitap_listesi):
    tur_dict = {}
    for k in kitap_listesi:
        tur_dict[k['tur']] = tur_dict.get(k['tur'], 0) + k['satis']
    return tur_dict
```

```python
# 📌 2. Liste ve Küme İşlemleri
turler = set(k['tur'] for k in kitaplar)
satis_1000_fazla = [k['isim'] for k in kitaplar if k['satis'] >= 1000]
```

```python
# 📌 3. Lambda / Filter / Map / Sorted
kitap_2020_sonrasi = list(map(lambda k: k['isim'], filter(lambda k: k['yil'] >= 2020, kitaplar)))
satis_10_artis = list(map(lambda k: round(k['satis']*1.10,2), kitaplar))
performans_siralamasi = sorted(kitaplar, key=lambda k: k['satis'], reverse=True)
```

```python
# 📌 4. Genel İstatistiksel Analiz
satislar = [k["satis"] for k in kitaplar]
ortalama_satis = statistics.mean(satislar)
std_sapma = statistics.stdev(satislar)
tur_satis = tur_satislari(kitaplar)
en_cok_satan_tur = max(tur_satis, key=tur_satis.get)
```

```python
# 📌 5. Train/Test (%70 train, %30 test)
random.seed(42)  
kitaplar_shuffle = kitaplar.copy()
random.shuffle(kitaplar_shuffle)

train_size = int(0.7 * len(kitaplar_shuffle))
train_set = kitaplar_shuffle[:train_size]
test_set = kitaplar_shuffle[train_size:]

# Train analizleri
train_satislar = [k['satis'] for k in train_set]
train_ortalama = statistics.mean(train_satislar)
train_std = statistics.stdev(train_satislar)
train_yazar = yazar_satislari(train_set)
train_tur = tur_satislari(train_set)

# Test analizleri
test_satislar = [k['satis'] for k in test_set]
test_ortalama = statistics.mean(test_satislar)
test_std = statistics.stdev(test_satislar) if len(test_satislar) > 1 else 0
test_yazar = yazar_satislari(test_set)
test_tur = tur_satislari(test_set)

# Karşılaştırma
genel_ortalama = statistics.mean(list(train_yazar.values()))
test_ust_ortalama = [k['isim'] for k in test_set if k['satis'] > genel_ortalama]
```

```python
# 📌 Analiz Sonuçları
print("~ ~ ~ ~ ~ KİTAP SATIŞ ANALİZ SİSTEMİ ~ ~ ~ ~ ~\n")

print("1. TEMEL ANALİZ")
print("----------------------------------------")
print("En çok satan kitap:", en_cok_satan(kitaplar)['isim'], f"({en_cok_satan(kitaplar)['satis']} adet)")

print("\nYazar satışları:")
for yazar, satis in yazar_satislari(kitaplar).items():
    print("*", yazar, ":", satis)

print("\nKitap türleri:")
for tur in turler:
    print("*", tur)

print("\n1000+ satış yapan kitaplar:")
for k in satis_1000_fazla:
    print("*", k)

print("\n2. FONKSİYONEL ANALİZ")
print("----------------------------------------")
print("2020 sonrası kitaplar:")
for k in kitap_2020_sonrasi:
    print("*", k)

print("\n%10 artış:")
for s in satis_10_artis:
    print("*", s)

print("\nPerformans sıralaması:")
for idx, k in enumerate(performans_siralamasi, 1):
    print(f"* {idx}. {k['isim']}: {k['satis']} adet")

print("\n3. İSTATİSTİKSEL ANALİZ")
print("----------------------------------------")
print("Ortalama satış:", round(ortalama_satis,2))
print("Standart sapma:", round(std_sapma,2))
print("En çok satan tür:", en_cok_satan_tur, f"({tur_satis[en_cok_satan_tur]} toplam satış)")

print("\n4. TRAIN/TEST")
print("----------------------------------------")
print("Train set ortalama satış:", round(train_ortalama,2))
print("Train set standart sapma:", round(train_std,2))

print("\nTrain set yazar satışları:")
for yazar, satis in train_yazar.items():
    print("*", yazar, ":", satis)

print("\nTrain set tür dağılımı:")
for tur, satis in train_tur.items():
    print("*", tur, ":", satis)

print("\nTest set ortalama satış:", round(test_ortalama,2))
print("Test set standart sapma:", round(test_std,2))

print("\nTest set yazar satışları:")
for yazar, satis in test_yazar.items():
    print("*", yazar, ":", satis)

print("\nTest set tür dağılımı:")
for tur, satis in test_tur.items():
    print("*", tur, ":", satis)

print("\nKarşılaştırmalı Sonuçlar")
print("----------------------------------------")
print("Genel ortalama:", round(genel_ortalama,2))
print("\nTest setinde ortalamanın üzerinde satış yapan kitaplar:")
for k in test_ust_ortalama:
    print("*", k)

# Kısa yorum
print("\nYorum")
print("----------------------------------------")
if test_ortalama > train_ortalama:
    print("* Test setinin ortalama satışları, train setinden daha yüksek çıkmıştır. Modelin genellenebilirliği olumlu görünüyor.")
else:
    print("* Train setinin ortalama satışları daha yüksek. Test setinde performans düşmüş olabilir.")
print("* Tür ve yazar dağılımlarında benzerlik/dengesizlik gözlenerek daha derin analiz yapılabilir.")
```
